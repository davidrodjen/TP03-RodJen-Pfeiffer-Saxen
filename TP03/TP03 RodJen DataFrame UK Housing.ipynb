```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
# Calling Dataset
data = pd.read_csv('price_paid_records.csv', encoding='latin')
```

```python
#Calling Specifically The Year 2016 & 2017
DataUK = data[data['Date of Transfer'].str.contains('2017')]
DataUK
```

```python
DataUK.info()
```

```python
#Slicing the Data to extract a specific Town
eastleigh_data = DataUK[DataUK['Town/City'] == 'EASTLEIGH'].copy()
eastleigh_data
```

```python
hampshire_data = DataUK[DataUK['County'] == 'HAMPSHIRE'].copy()
hampshire_data
```

```python
# Isolate the cost of the homes in Hampsire based on districts
groupByHampsire = hampshire_data['Price'].groupby(DataUK['District'])
groupByHampsire.mean().plot.bar()
```

```python
# Isolate the Greater London Price points
greaterLondon_data = DataUK[DataUK['County'] == 'GREATER LONDON'].copy()
greaterLondon_data
```

```python
# Look at the mean price by district in the Greater London County
groupByLondon = greaterLondon_data['Price'].groupby(DataUK['District'])
groupByLondon.mean().plot.bar()
```
