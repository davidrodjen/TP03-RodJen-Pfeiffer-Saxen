```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ldn = pd.read_csv('London.csv', encoding = 'latin')
ldn.head(6)
```

```python
ldnHouse = ldn[ldn['House Type'].str.contains('House')]
ldnHouse
```

```python
plt.scatter(ldnHouse['Price'], ldnHouse['Area in sq ft'])
plt.title('Home Prices in London by Square Footage')
plt.xlabel('Price in Millions of Pounds')
plt.ylabel('Square Feet')
plt.show()
```

```python
ldnHomesUnderOneMil = ldn[ldn['Price'] < 1000000]
ldnHomesUnderOneMil
```

```python
plt.scatter(ldnHomesUnderOneMil['Price'], ldnHomesUnderOneMil['Area in sq ft'])
plt.title('Home Prices in London by Square Footage')
plt.xlabel('Price in Millions of Pounds')
plt.ylabel('Square Feet')
plt.show()
```

```python
aptInldn = ldn[ldn['House Type'].str.contains('Flat / Apartment')]
aptInldn
```

```python
plt.scatter(aptInldn['Price'], aptInldn['Area in sq ft'])
plt.title('Flat Prices in London by Square Footage')
plt.xlabel('Price in Millions of Pounds')
plt.ylabel('Square Feet')
plt.show()
```

```python
avgPrice = ldn['Price'].mean()
avgSqFt = ldn['Area in sq ft'].mean()
print ("The average home Price in London is: £" ,int(avgPrice))
print ("The average square footage of a home in London is :", int(avgSqFt), "Square Feet")
```
