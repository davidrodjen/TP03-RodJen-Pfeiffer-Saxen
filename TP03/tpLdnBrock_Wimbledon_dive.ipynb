```python
unneeded_columns = ['Unnamed: 0', 'Property Name', 'No. of Bedrooms', 'No. of Bathrooms', 'No. of Receptions', 'City/County', 'Postal Code']
WimbledonHouses2 = ldn[(ldn['House Type'] == 'House') & (ldn['Location'] == 'Wimbledon')].copy()
WimbledonHouses2 = WimbledonHouses2.drop(columns=unneeded_columns)

quartile = WimbledonHouses2['Area in sq ft'].quantile([0.001,.20, 0.40, 0.60, 0.80, 1])

# Define bin labels
bin_label = ['Low', 'Low Medium','Medium', 'High', 'Very High']

# Create a new column with the quartile bin labels
WimbledonHouses2['quartile_bins'] = pd.cut(WimbledonHouses2['Area in sq ft'], bins=quartile, labels=bin_label)

# Display the updated DataFrame
print(WimbledonHouses2)

print(WimbledonHouses2[(WimbledonHouses2['quartile_bins'] == 'Q1')])
```

```python
#Finding Mix/Max per bin

bin_ranges = WimbledonHouses2.groupby('quartile_bins')['Area in sq ft'].agg(['min', 'max'])
bin_ranges
```

```python
# Creating re-usable function that can be utilized in different bin like "Q1" etc
def get_bin_info(df, bin_label):
    
    #Filtering by bin
    bin_df = df[df['quartile_bins'] == bin_label]
    
    # Finding the average sqft, mode, min, max, and avg total cost
    item_count = len(bin_df)
    avg_sqft = bin_df['Area in sq ft'].mean()
    mode = bin_df['Area in sq ft'].mode().iloc[0]
    min_sqft = bin_df['Area in sq ft'].min()
    max_sqft = bin_df['Area in sq ft'].max()
    avg_total_cost = bin_df['Price'].mean()
    avg_price_per_sqft = avg_total_cost / avg_sqft

    
    # Returning as a dictionary
    bin_info = {
        'count': item_count,
        'average_square_feet': avg_sqft,
        'mode': mode,
        'min_sqft': min_sqft,
        'max_sqft': max_sqft,
        'average_total_cost': avg_total_cost,
        'avg_price_per_sqft': avg_price_per_sqft
    }
    return bin_info
```

```python
# Values for  Wimbledon housing that is in the 'Low' category
Wimbledon_Low_Housing_General_information = get_bin_info(WimbledonHouses2,'Low')
print(Wimbledon_Low_Housing_General_information)
```

```python
Wimbledon_Medium_Housing_General_information = get_bin_info(WimbledonHouses2,'Medium')
print(Wimbledon_Medium_Housing_General_information)
```

```python
# Interesting to note that the higher you go up in category the better you get at per sqft
Wimbledon_Very_High_Housing_General_information = get_bin_info(WimbledonHouses2,'Very High')
print(Wimbledon_Very_High_Housing_General_information)
```

```python
# That was selecting a random location name I found in the list which was Wimbledon
# Now I want to identify the priciest location in London based on the dataset

# Group the DataFrame by 'Location' and calculate the mean price for each location
avg_price_by_location = ldn.groupby('Location')['Price'].mean()

# Find the location with the highest average price
highest_avg_price_location = avg_price_by_location.idxmax()

# Also findin the location with the lowest average price
lowest_avg_price_location = avg_price_by_location.idxmin()

print(highest_avg_price_location)
print(lowest_avg_price_location)
```
